```python
import bs4
from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime
```

```python
url='https://weather.gc.ca/city/pages/nl-24_metric_e.html'
page = requests.get(url)
print(page)
soup =  BeautifulSoup(page.content,'html.parser')
print(soup)
```

```python
table = soup.find_all('table')
```

```python
raw_data = [row.text.splitlines() for row in table]
raw_data = raw_data[:1000]
len(raw_data)
```

```python

print(raw_data)
```

```python
data = pd.DataFrame(columns=["Today","Today_for" "Tomorrow","Tomorrow_fore","Third","Third_fore","Fourth","Fourth_fore","Fifth","Fifth_fore","Six","Sith_fore","Seventh","Seventh_fore"])
data

for row in soup.find_all('tbody'):
    col = row.find_all('td')
    if (col != []):
        
        Today = col[0].text.strip()
        Today_fore = col[1].text.strip()
        Tomorrow = col[2].text.strip()
        Tomorrow_fore = col[3].text.strip()
        Third = col[4].text.strip()
        Third_fore = col[5].text.strip()
        Fourth = col[6].text.strip()
        Fourth_fore = col[7].text.strip()
        Fifth = col[8].text.strip()
        Fifth_fore = col[9].text.strip()
        Sixth = col[10].text.strip()
        Sixth_fore = col[11].text.strip()
        Seventh = col[12].text.strip()
        Seventh_fore = col[13].text.strip()
        data = data.append({"Today":Today,"Today_fore":Today_fore,"Tomorrow":Tomorrow,"Tomorrow_fore":Tomorrow_fore,"Third":Third,"Third_fore":Third_fore,"Fourth":Fourth,"Fourth_fore":Fourth_fore,"Fifth":Fifth,"Fifth_fore":Fifth_fore,"Sixth":Sixth,"Sixth_fore":Sixth_fore,"Seventh":Seventh,"Seventh_fore":Seventh_fore }, ignore_index=True)
```
